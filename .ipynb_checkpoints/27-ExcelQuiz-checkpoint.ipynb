```python
#2) Bu firma ortalama ne kadar maaş vermektedir?
```

```python
average_salary = dataFrame["Salary"].mean()
average_salary
```

```python
#3) Bu firmada departmanlara göre ortalama maaş karşılaştırması nasıldır?
```

```python
departmentGroup = dataFrame.groupby("Department")["Salary"]
departmentGroup.mean()
```

```python
#4) Bu firmada title (senior - junior) durumuna göre ortalama maaş karşılaştırması nasıldır?
```

```python
titleGroup = dataFrame.groupby("Title")["Salary"]
titleGroup.mean()
```
